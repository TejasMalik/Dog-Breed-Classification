```python
import os


def move_file(old_file_path, new_directory):
    if not os.path.isdir(new_directory):
        os.mkdir(new_directory)
    base_name = os.path.basename(old_file_path)
    new_file_path = os.path.join(new_directory, base_name)
    # Deletes a file if that file already exists there, you can change this behavior
    if os.path.exists(new_file_path):
        os.remove(new_file_path)
    os.rename(old_file_path, new_file_path)
```

```python
cop_folder = 'train\\'
destination_folder = 'data\\'
df = pd.read_csv('labels.csv', header=None, skiprows=[0])
```

```python
images = list(df[0])
breed = list(df[1])
```

```python
for i in range(len(images)):
    if (breed[i] == 'beagle' or breed[i] == 'chihuahua' or
        breed[i] == 'doberman' or breed[i] == 'french_bulldog' or 
        breed[i] == 'golden_retriever' or breed[i] == 'malamute' or 
        breed[i] == 'pug' or breed[i] == 'saint_bernard' or 
        breed[i] == 'scottish_deerhound' or breed[i] == 'tibetan_mastiff'):
        filename = os.path.join(cop_folder, images[i]+'.jpg')
        move_file(filename, destination_folder)
        
```

```python
targets = {}
for i in range(len(images)):
    if (breed[i] == 'beagle' or breed[i] == 'chihuahua' or
        breed[i] == 'doberman' or breed[i] == 'french_bulldog' or 
        breed[i] == 'golden_retriever' or breed[i] == 'malamute' or 
        breed[i] == 'pug' or breed[i] == 'saint_bernard' or 
        breed[i] == 'scottish_deerhound' or breed[i] == 'tibetan_mastiff'):
        targets[images[i]] = breed[i]
```

```python
pd.to_csv(targets)
```

```python
dataset = pd.DataFrame({'id': list(targets.keys()), 'breed': list(targets.values())})
```
